## Libraries and Data 

```python
import pandas as pd
from datetime import timedelta
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("online_shoppers.csv",
                encoding = "ISO-8859-1").dropna()
df.head()
```

## Data Preparation 

```python
#Creating sales column
df['sales'] = df['Quantity'] * df['UnitPrice']
df.head(4)
```

```python
#Dates - Transform Date variable
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

#Get Last date available
snapshot_date = df['InvoiceDate'].max() + timedelta(days = 1)
snapshot_date
```

```python
#Aggregate on Customer Level

customer = df.groupby(['CustomerID']).agg({
    #Recency
    #x.max() is the last day of the customer
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'count',
    'sales': 'sum'
})
customer.head()
```

```python
#Create Basket / Monetary Variable

customer['Monetary'] = customer.sales / customer.InvoiceNo
customer.head()
```

```python
#Dropping Sales Variable
df = customer.drop(columns='sales')
df.head(1)
```

```python
#Changing Variable Names
df.rename(columns = {'InvoiceDate': 'Recency',
                    'InvoiceNo': 'Frequency'}, 
                      inplace = True)

df.head(1)
```

## RFM Preparation 

```python
#Create Frequency Groups
df['F'] = pd.qcut(x = df['Frequency'], q = 4, labels = range(1,5))
df['M'] = pd.qcut(x = df['Monetary'], q = 4, labels = range(1,5))
df['R'] = pd.qcut(x = df['Recency'], q = 4, labels = range(4, 0, -1))
df.head()
```

Higher Recency would mean lower value

```python
#RFM Score
df['RFM'] = df[['R', 'F', 'M']].sum(axis = 1)
df.head(2)
```

```python
#Creating the RFM function
def rfm_segment(df):
    if df['RFM'] >= 11:
        return "SuperStar"
    elif((df['RFM'] >= 8) and (df['RFM'] < 11)):
        return "Future Champion"
    elif((df['RFM'] >= 6) and (df['RFM'] < 8)):
        return 'High Potential'
    else:
        return 'Low Relevance' 
```

## RFM

```python
#Applying the RFM Function
df['RFM_level'] = df.apply(rfm_segment, axis = 1)
df.head()
```

```python
#Looking into the segments
df.groupby('RFM_level').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
})
```
